# f)

````python
df = df.astype({'mpg': 'int', 'displacement': 'int', 'weight': 'int'})

lista = ['displacement', 'weight']

fig, axs = plt.subplots(1, 2)
axs = axs.flatten()

for idx, values in enumerate(lista):
    mymodel = np.poly1d(np.polyfit(df['mpg'], df[values], 3))

    myline = np.linspace(df['mpg'].max(), df['mpg'].min(), df[values].max())

    sns.scatterplot(data=df, x='mpg', y=values, ax=axs[idx])
    axs[idx].plot(myline, mymodel(myline), color='red')
    axs[idx].set_title(f'mpg X {values}')

plt.tight_layout()
plt.show()
````

Aqui podemos perceber que tanto Displacement como Weight tem uma relação com Mpg. Ao plotar eu descobri, visualmente, que eles pareciam ter uma relação Polinomial. Para melhor representar fiz uma Linha de uma regressão polinomial

````python
mean_all = df.mean()

data_with_min_medv = df[df['medv'] == df['medv'].min()]
mean_data_with_medv = data_with_min_medv.mean()

mean_comparison = pd.DataFrame({'Média de todos os parâmetros': mean_all, 'Média dos parâmetros quando os subúrbios tem o menor medv': mean_data_with_medv})

mean_comparison.drop(['medv', 'suburb'], errors='ignore', inplace=True)

# Plot
mean_comparison.plot(kind='bar', figsize=(12,6))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

Aqui podemos ver que a média dos subúrbios com menor **Medv** nos mostra certas coisas:

+ A média da criminalidade nesses lugares é maior
+ a média da Taxa nesses lugares é maior
+ Em outros parâmetros é maior também mas não aparenta ser muito significativo

````python
more_than_7_rooms = df[df['rm'] > 7].shape

more_than_8_rooms = df[df['rm'] > 8].shape

print(f'O número de subúrbios com mais de 7 quartos é : {more_than_7_rooms}')

print(f'O número de subúrbios com mais de 8 quartos é : {more_than_8_rooms}')

data_with_rm_more_than_8_rooms = df[df['rm'] > 8].mean()

mean_comparison = pd.DataFrame({'Média dos subúrbios com qualquer quantidade de quartos': mean_all, 'Média dos subúrbios com mais de 8 quartos': data_with_rm_more_than_8_rooms})

mean_comparison.drop(['rm', 'suburb'], errors='ignore', inplace=True)

# Plot

mean_comparison.plot(kind='bar', figsize=(12,6))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````
