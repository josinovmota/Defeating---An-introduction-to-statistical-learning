# b)

```python
# O exerc√≠cio pede pra prevermos o sal√°rio de um College Graduate com 110 de QI e 4 de GPA
def predict_salary(Qi: int, GPA: int, Type: int):

    B_hat_QI: float = 0.07
    B_hat_GPA: int = 20
    B_hat_3: int = 35
    B_hat_4: float = 0.01
    B_hat_5: int = -10
    B_hat_0: int = 50

    Sallary: int = 1000

# Importante ressaltar que todos os predictors devem ser multiplicados pelo B_hat value
    Final_Value = ((B_hat_QI * Qi + B_hat_GPA * GPA + B_hat_3 * Type + B_hat_4 * (Qi*GPA) + B_hat_5 * (Type*GPA) + B_hat_0) * Sallary)
    Formated_Final_Value = "R$ {:.0f}".format(Final_Value)
    return(print(Formated_Final_Value))

predict_salary(110, 4, 1)
```

# h)

```python
# Atribuindo model em results para facilitar a cria√ß√£o do c√≥digo
results = model

# Pegando results e residuals
residuals = results.resid
leverage = results.get_influence().hat_matrix_diag

# Todo o c√°lculo que √© feito dos outliers e leverage points baseados em um treshold de 3 e 2, respectivamente
outlier_threshold = np.abs(residuals).mean() + 3 * np.abs(residuals).std()
leverage_threshold = 2*(x_train.shape[1])/x_train.shape[0]

# Capturando apenas aqueles que s√£o maiores que o threshold, ou seja, os respectivos outliers e high leverage
# points
outlier_indices = np.where(np.abs(residuals) > outlier_threshold)
high_leverage_indices = np.where(leverage > leverage_threshold)

# Printando os outliers e high leverage points dado um treshold
print('Outliers:', outlier_indices)
print('High-leverage points:', high_leverage_indices)
```
