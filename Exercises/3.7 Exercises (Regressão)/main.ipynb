# b)

```python
# O exerc√≠cio pede pra prevermos o sal√°rio de um College Graduate com 110 de QI e 4 de GPA
def predict_salary(Qi: int, GPA: int, Type: int):

    B_hat_QI: float = 0.07
    B_hat_GPA: int = 20
    B_hat_3: int = 35
    B_hat_4: float = 0.01
    B_hat_5: int = -10
    B_hat_0: int = 50

    Sallary: int = 1000

# Importante ressaltar que todos os predictors devem ser multiplicados pelo B_hat value
    Final_Value = ((B_hat_QI * Qi + B_hat_GPA * GPA + B_hat_3 * Type + B_hat_4 * (Qi*GPA) + B_hat_5 * (Type*GPA) + B_hat_0) * Sallary)
    Formated_Final_Value = "R$ {:.0f}".format(Final_Value)
    return(print(Formated_Final_Value))

predict_salary(110, 4, 1)
```

```python
# Importa√ß√£o das libs 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import os

from sklearn.linear_model import LinearRegression
```

```python
# Obten√ß√£o dos dados
cwd = os.getcwd()
df = pd.read_csv(cwd + '\\..\\..\\Data\\Auto.csv')

# Defini√ß√£o do X e Y
x_train = df['mpg']
y_train = df['horsepower']

# Convertendo os valores de Y_train pra num√©ricos, pois apareceu um "?" ali no meio e a Series inteira virou object
y_train = pd.to_numeric(y_train, errors='coerce')
# Transformar os "?" na m√©dia de todos os valores
y_train.fillna(y_train.mean(), inplace=True)
# Adicionar uma constant no X_train
x_train = sm.add_constant(x_train)

# Fit no modelo
model = sm.OLS(y_train, x_train).fit()  

# Summary
print(model.summary())
```
