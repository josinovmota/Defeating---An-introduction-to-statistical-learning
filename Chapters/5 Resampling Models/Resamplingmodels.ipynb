# Resampling models

Tirar e colocar dados de uma sample é chamado de Resampling. Ele é usado de maneira similar 
aos métodos aparados. Imagine que busquemos a média da altura de alunos real usando uma sample. Caso façamos a média com apenas 5 pessoas, podemos acabar pegando os 5 mais altos
ou os 5 mais baixos, nos fazendo assim ter uma média totalmente enviesada. Fazer o resampling nos permite ter uma ideia melhor da distribuição dos dados.

## Cross-Validation:

Sempre que vamos criar um modelo, precisamos analisar os dados afim de entender qual modelo se encaixa melhor. De início, usamos dados de treino para treinar o nosso modelo e, à partir das predictions feitas nesses dados de treino, fazemos as seguintes. Isso faz com que nosso modelo seja bom para dos dados de treino, ou seja, tenha um BAIXO TRAINING ERROR RATE, porém, pode ser que os dados que venham a seguir não tenham as mesmas característica dos de treino, tornando assim, muitas vezes, nosso modelo pior para os dados de TESTE, causando um alto TEST ERROR RATING.
Dado isso, a cross-validation divide os dados de treino em training set e **VALIDATION SET** e faz resampling com uma das partes divididas, fazendo assim, com que tenhamos certeza que nosso modelo tem um bom desempenho mesmo com a variabilidade dos dados

### Leave-One-Out Cross-Validation:

É um dos métodos de cross-validation aonde, ao invés de dividir nossos dados em TRAINING SET e VALIDATION SET, refazemos nosso modelo sempre retirando 1 AMOSTRA e prevendo ela de forma singular. Ele requer mais tempo de execução na maioria dos casos, porém, tende a lidar muito bem com BIAS, já que eventualmente usará como TREINO cada um dos x e y.

### K-Fold Cross-Validation:

Um dos motivos para usarmos o K-Fold CV no lugar do LOOCV é que o K-Fold requer menos tempo, ou seja, menos trabalho computacional. Em alguns tipos de regressões existe uma fórmula que diminui drasticamente esse tempo, porém, em grandes modelos, fazer 1 por 1, como é o caso do LOOCV, requer muito tempo. Então, por mais que o Leave-One-Out possa gerar algo mais preciso, ele sempre caminha ao lado do K-Fold.

O K dentro do K-Fold significa a quantidade de grupos, ou seja, se o K = 10, dividimos os dados em 10 grupos de tamanho igual

Quando plotamos os métodos de Cross-Validation em relação à flexibilidade, sabemos que vai existir um ponto mínimo, aonde o MSE é o menor possível e logo após esse ponto, o MSE vai começar a subir. #### Bias-Variance Trade-Off for k-fold cross-validation:

O K-Fold tende a ser, muitas vezes, melhor que o LOOCV tanto pelo tempo computacional, como por dar resultados melhores que o LOOCV. Isso é explicado pelo Bias-Variance Trade-Off de ambos
