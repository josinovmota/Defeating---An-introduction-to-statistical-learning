# Resampling models

Tirar e colocar dados de uma sample é chamado de Resampling. Ele é usado de maneira similar 
aos métodos aparados. Imagine que busquemos a média da altura de alunos real usando uma sample. Caso façamos a média com apenas 5 pessoas, podemos acabar pegando os 5 mais altos
ou os 5 mais baixos, nos fazendo assim ter uma média totalmente enviesada. Fazer o resampling nos permite ter uma ideia melhor da distribuição dos dados.

## Cross-Validation:

Sempre que vamos criar um modelo, precisamos analisar os dados afim de entender qual modelo se encaixa melhor. De início, usamos dados de treino para treinar o nosso modelo e, à partir das predictions feitas nesses dados de treino, fazemos uma prediction com os dados de teste. Isso faz com que nosso modelo seja bom para dos dados de treino, ou seja, tenha um BAIXO TRAINING ERROR RATE, porém, pode ser que os dados que venham a seguir não tenham as mesmas característica dos de treino, tornando assim, muitas vezes, nosso modelo pior para os dados de TESTE, causando um alto TEST ERROR RATING.
Dado isso, a cross-validation divide os dados de treino em training set e **VALIDATION SET** e faz resampling com uma das partes divididas, fazendo assim, com que tenhamos certeza que nosso modelo tem um bom desempenho mesmo com a variabilidade dos dados

### Leave-One-Out Cross-Validation:

É um dos métodos de cross-validation aonde, ao invés de dividir nossos dados em TRAINING SET e VALIDATION SET, refazemos nosso modelo sempre retirando 1 AMOSTRA e prevendo ela de forma singular. Ele requer mais tempo de execução na maioria dos casos, porém, tende a lidar muito bem com **BIAS**, já que eventualmente usará como TREINO cada um dos x e y.

### K-Fold Cross-Validation:

Um dos motivos para usarmos o K-Fold CV no lugar do LOOCV é que o K-Fold requer menos tempo, ou seja, menos trabalho computacional. Em alguns tipos de regressões existe uma fórmula que diminui drasticamente esse tempo, porém, em grandes modelos, fazer 1 por 1, como é o caso do LOOCV, requer muito tempo. Então, por mais que o Leave-One-Out possa gerar algo mais preciso, ele sempre caminha ao lado do K-Fold.

O K dentro do K-Fold significa a quantidade de grupos, ou seja, se o K = 10, dividimos os dados em 10 grupos de tamanho igual

Quando plotamos os métodos de Cross-Validation em relação à flexibilidade, sabemos que vai existir um ponto mínimo, aonde o MSE é o menor possível e logo após esse ponto, o MSE vai começar a subir. #### Bias-Variance Trade-Off for k-fold cross-validation:

K-fold CV tem uma vantagem computacional em relação ao LOOCV, pois o mesmo faz menos iterações dentro do **N**. Enquanto o LOOCV faz N iterações, o K-Fold CV faz K iterações aonde K < N

O Bias-Variance Trade-Off nesse caso nos mostra que o LOOCV tende a ter uma **VARIANCE** maior, porém um **BIAS** menor. É sabido que um modelo bom é aquele modelo que tem POUCA **VARIANCE** e POUCO **BIAS**, como é mostrado a seguir:



Sabendo disso e entendendo que o LOOCV nada mais é do que um K-Fold CV com K = N, e que quando o K < N existe uma tendência a ter um Maior **BIAS** e menos **VARIANCE**, procuramos chegar no meio-termo, ou seja, algo que não seja K = N (Afim de ter menos **VARIANCE**), e que seja menor do que N (Afim de ter menos **BIAS**). É dito que valores de K no K-Fold CV que tendem a ter menos **VARIANCE** e menos **BIAS** são K = 5 e K = 10

### Bootstrap

O processo de bootstrapping nos ajuda a calcular algumas estatísticas de datasets, como **Standard Error**, **Standard Deviation**, sem precisar de dados reais. Nesse processo fazemos um **Resampling** do dataset com replacement, ou seja, tiramos uma amostra dos dados, que podem ser repetidas, calculamos o que buscamos, criamos um novo bootstrap data set com os dados reais e repetimos esse processo várias vezes. À partir de vários testes, foi descoberto que esses cálculos por meio do bootstrapping dão valores muito próximos aos que calculamos usando datasets anteriores.
